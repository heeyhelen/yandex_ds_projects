```python
plt.figure(figsize=(15, 7))

plt.plot(taxi)
plt.plot(taxi.rolling(24).mean(), color='green')
plt.plot(taxi.rolling(24).std(), color='purple')

plt.title('Распределение заказов в час со скользящим стандартным отклонением каждые 24 часа', fontsize=17)
plt.legend(['Кол-во заказов в час', 'Скользящее среднее каждые 24 часа',
            'Скользящее стандартное отклонение каждые 24 часа', 'stn'])
plt.xlabel('Дата', fontsize=13)
plt.ylabel('Количество заказов', fontsize=13)
plt.show()
```

Примерно до середины июня временной ряд выглядит стационарным, потом кол-во заказов, а вместе с ним и скользящее среднее, и скользящее стандартное отклонение растут, т.е. ряд нестационарен. 

Проведем Augmented Dickey-Fuller тест. 
H0 = временной ряд имеет единичный корень, т.е. временной ряд нестационарен.

```python
def ADF_test(timeseries):
  
    result = adfuller(timeseries, autolag='AIC')
    
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
        
    if result[0] > result[4]['5%']:
        print("Временной ряд скорее нестационарен.")
    else:
        print("Временной ряд скорее стационарен.")
```

```python
ADF_test(taxi)
```

Статистика ADF составляет -3,0689, а соответствующее p-value равно 0,0289, что меньше уровня значимости 5% (0,05). Это означает, что мы можем отвергнуть нулевую гипотезу и сделать вывод, что ряд, вероятно, является стационарным.

## Обучение

```python
def RMSE(target, predict):
    RMSE = math.sqrt(mean_squared_error(target, predict))
    return RMSE
```

```python
def search_params(model, params):
    
    tscv = TimeSeriesSplit(n_splits=N)    
    best_rmse = 48
    best_model = 0
        
    gsearch = GridSearchCV(estimator=model,
                              param_grid=params,
                              scoring='neg_root_mean_squared_error',
                              n_jobs=-1,
                              cv=tscv, 
                              verbose=1)
        
    gsearch.fit(features_train, target_train)
            
    print('Лучшие параметры:', gsearch.best_params_)
    print('Лучший rmse:', gsearch.best_score_ * (-1))
    
    model = gsearch.best_estimator_
    rmse = gsearch.best_score_ * (-1)
    
    return model, rmse 
```

```python
def plot_pred_train(model, name):    
    
    predicted_df = pd.DataFrame(model.predict(features_train), index=features_train.index)
    
    plt.figure(figsize=(15,7))
    
    plt.plot(target_train['2018-07-27':'2018-08-02'], color='black', alpha=0.7)
    plt.plot(predicted_df['2018-07-27':'2018-08-02'], color='red', alpha=0.7)
    
    plt.title(f'{name}: cравнение предсказаний на train выборке, неделя', fontsize=17)
    plt.ylabel('Количество заказов', fontsize=13)
    plt.xlabel('Дата', fontsize=13)
    plt.legend(['Реальные', 'Предсказанные'])
    plt.ylim((0, 250))
    
    return plt.show()
```

```python
def make_features(data, max_lag, rolling_mean_size):

    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```
